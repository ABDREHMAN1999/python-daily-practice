### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here

heads = int(input("Enter the number of heads: "))
legs =  int(input("Enter the number of legs: "))
from sympy import symbols , Eq , solve
#unknown variables
D , C = symbols('D, C')
eq1 = Eq((4*D+2*C), legs)
eq2 = Eq((D + C), heads)
print("number of cats and dogs are: ")
result = (solve((eq1,eq2),(D,C)))
print("chicken -> ",result[C])
print("dogs -> ",result[D])



````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter the value for n: "))
result = 0
for i in range(0, n+1):
  result = result+(i**2)
print(result)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
a = float(input("Enter the first term: "))
b = float(input("Enter the second term: "))
d = b - a  #this is the common diffrence
n = int(input("Enter the value of n: "))
nth_term = a+(n-1)*d
print("The nth term is :",nth_term )
````
